```python
no_times_scarlet_predicted = 0
ins_count = 0
value_count = 0
correct_count = 0
wrong_count = 0
val_check = ["Not Specified","",None]
blank_values_scarlet_and_carl = 0
sheet2 = pd.DataFrame()
sheet1 = pd.DataFrame()
value_count_dict = {}
no_f_t_scar_pre = {}
flag = 0

for x in tra_data['result']:
    file_path = x['contractInfo']['contract_path'] + '/' + x['contractInfo']['file_name']
    data = json.dumps({"file_path":file_path, "datamodel_structure":datamodel_structure, "datamodel_id":datamodel_id
           })
    
    response = requests.post(url, data=data)
    response = response.json()
    
#     if flag == 0:
#         for value_m_k in response['result'][0].keys():
#             for value_innr_k in response['result'][0][value_m_k].keys():
#                 temp = {value_innr_k:0}
#                 value_count_dict.update(temp)
#         flag = 1
#     else:
#         pass

    for main_k in response['result'][0].keys():
        
        if main_k not in no_f_t_scar_pre.keys():
            temp_m = {main_k:{}}
            no_f_t_scar_pre.update(temp_m)
        else:
            pass
        
        for innr_k in response['result'][0][main_k].keys():
            
            # value count
            if innr_k in value_count_dict.keys():
                value_k = value_count_dict[innr_k]
                value_count_dict[innr_k] = value_k + 1
            else:
                temp1 = {innr_k:1}
                value_count_dict.update(temp1)
            
            # starting inner loop of key 
            for in_innr_k in response['result'][0][main_k][innr_k].keys():
                
                # value
                if in_innr_k == 'value':
                    scarlet_val = response['result'][0][main_k][innr_k][in_innr_k]
                    
                    if innr_k in no_f_t_scar_pre[main_k].keys():
                        if 'no_times_scarlet_predicted' in no_f_t_scar_pre[main_k][innr_k].keys():
                            val_pre = no_f_t_scar_pre[main_k][innr_k]['no_times_scarlet_predicted']

                            # no_times_scarlet_predicted                    
                            if scarlet_val not in val_check:
                                no_f_t_scar_pre[main_k][innr_k]['no_times_scarlet_predicted'] = val_pre + 1
                            else:
                                pass
                                # no_times_scarlet_predicted = no_times_scarlet_predicted + 1
                        else:
                            pass
                    else:
                        tempee = {innr_k:{'no_times_scarlet_predicted':1}}
                        no_f_t_scar_pre[main_k].update(tempee)
                    
                    
                    carl_val = tra_data['result'][0]['keyValues']['datamodel_structure'][0][main_k][innr_k][in_innr_k]
                    
                    
                    # wrong and blank check
                    if 'correct_count' and 'wrong_count' and 'blank_values_scarlet_and_carl' in no_f_t_scar_pre[main_k][innr_k].keys():
                        correct_count = no_f_t_scar_pre[main_k][innr_k]['correct_count']
                        wrong_count = no_f_t_scar_pre[main_k][innr_k]['wrong_count']
                        blank_values_scarlet_and_carl = no_f_t_scar_pre[main_k][innr_k]['blank_values_scarlet_and_carl']
                    else:
                        co_n_wr_c = {'correct_count':0, 'wrong_count':0, 'blank_values_scarlet_and_carl':0}
                        no_f_t_scar_pre[main_k][innr_k].update(co_n_wr_c)
                    
                    if scarlet_val == '' and carl_val == '':
                        no_f_t_scar_pre[main_k][innr_k]['blank_values_scarlet_and_carl'] = blank_values_scarlet_and_carl + 1
                    elif scarlet_val == carl_val:
                        no_f_t_scar_pre[main_k][innr_k]['correct_count'] = correct_count + 1
                    else:
                        no_f_t_scar_pre[main_k][innr_k]['wrong_count'] = wrong_count + 1
                else:
                    pass


                # instruction
                if in_innr_k == 'instruction':
                    carl_instruction = tra_data['result'][0]['keyValues']['datamodel_structure'][0][main_k][innr_k][in_innr_k] 
                    ins_vals = response['result'][0][main_k][innr_k][in_innr_k]
                    if ins_vals == "":
                        ins_count = ins_count + 0
                    else:
                        ins_count = ins_count + 1
                else:
                    pass



                # values for sheet 2
                contract_name = x['contractInfo']['contract_name']
                file_name = x['contractInfo']['file_name']
            
            if scarlet_val != '' and carl_val != '':
                temp2 = pd.DataFrame({'contract_name': [contract_name], 'file_name': [file_name], 
                                      'header':[main_k], 'key_name':[innr_k],
                                      'carl_val': [carl_val], 'carl_instruction':[carl_instruction], 
                                      'scarlet_value':[scarlet_val]})
                sheet2 = pd.concat([sheet2, temp2])
            else:
                pass
        
        
            # find type
            if 'type' not in no_f_t_scar_pre[main_k][innr_k].keys():
                for x_type in range(len(data_model_json['result'][0]['datamodel_structure'])):
                    if data_model_json['result'][0]['datamodel_structure'][x_type]['header_name'] == main_k:
                        for y_type in range(len(data_model_json['result'][0]['datamodel_structure'][x_type]['fields'])):
                            if data_model_json['result'][0]['datamodel_structure'][x_type]['fields'][y_type]['label_name'] == innr_k:
                                typ = data_model_json['result'][0]['datamodel_structure'][x_type]['fields'][y_type]['type']
                                typ_temp = {'type':typ}
                                no_f_t_scar_pre[main_k][innr_k].update(typ_temp)
                            else:
                                pass
                        pass
            else:
                pass
```
